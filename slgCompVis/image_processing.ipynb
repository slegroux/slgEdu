# Correlation

- image kernels explained visually
http://setosa.io/ev/image-kernels/

```python
sharpen = np.array(
            [[0,-1,0],
            [-1,5,-1],
            [0,-1,0]])
plt.imshow(sharpen, cmap='gray', interpolation='None')
```

```python
top = np.array(
            [[-1,-1,-1],
            [1,1,1],
            [0,0,0]])
plt.imshow(top, cmap='gray', interpolation='None')
```

```python
from skimage import color
x1 = 88
x2 = 166
y1 = 132
y2 = 162
gray_crop = color.rgb2gray(crop[y1:y2,x1:x2])
print np.shape(gray_crop)
plt.imshow(gray_crop, cmap='gray')
```

```python
from scipy.ndimage import filters
fig, axes = plt.subplots(2,1,sharey=True)
corr_sharp = filters.correlate(gray_crop, sharpen)
corr_top = filters.correlate(gray_crop, top)
axes[0].imshow(corr_sharp,cmap='gray')
axes[1].imshow(corr_top,cmap='gray')
```

# Convolution

```python
filt = np.rot90(top,2)
plt.imshow(filt,cmap='gray', interpolation='None')
```

```python
fig, axes = plt.subplots(2,1, sharey=True)
corr_top = filters.correlate(gray_crop, top)
conv_top = filters.convolve(gray_crop, filt)
axes[0].imshow(corr_top,cmap='gray')
axes[1].imshow(conv_top,cmap='gray')
```

# Pooling

```python
from skimage.measure import block_reduce
def pool(im): return block_reduce(im, (8,8), np.max)
pooled = pool(corr_top)
plt.imshow(pooled,cmap='gray',interpolation='None')
```

# Edge detection

```python
from skimage.feature import canny
from skimage import morphology
```

```python
camera = data.camera()
edges = canny(camera)
```

```python
fig = plt.figure(figsize=(7,7))
plt.imshow(edges, cmap='gray')
```

```python
from scipy import ndimage as ndi
fig = plt.figure(figsize=(6,6))
fill_img = ndi.binary_fill_holes(edges)
plt.imshow(fill_img, cmap='gray')
```

```python
elevation_map = filters.sobel(camera)
fig = plt.figure(figsize=(5,5))
plt.imshow(elevation_map,cmap=plt.cm.jet,interpolation='nearest')
```

```python
markers = np.zeros_like(camera)
markers[camera < 30] = 2
markers[camera > 150] = 1

fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(markers, cmap=plt.cm.spectral, interpolation='nearest')
ax.axis('off')
ax.set_title('markers')
```

```python
segmentation = morphology.watershed(elevation_map, markers)

fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(segmentation, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('segmentation')
```

```python
from skimage.color import label2rgb
from scipy import ndimage

segmentation = ndimage.binary_fill_holes(segmentation - 1)
labeled_coins, _ = ndimage.label(segmentation)
image_label_overlay = label2rgb(labeled_coins, image=camera)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
ax1.imshow(camera, cmap=plt.cm.gray, interpolation='nearest')
ax1.contour(segmentation, [0.5], linewidths=1.2, colors='y')
ax1.axis('off')
ax2.imshow(image_label_overlay, interpolation='nearest')
ax2.axis('off')

# fig.subplots_adjust(**margins)
```
