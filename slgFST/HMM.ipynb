# HMM

```python
import openfst_python as fst
```

## Symbol Tables

Create symbol table that associates symbol to integer

```python
input_sym = fst.SymbolTable()
output_sym = fst.SymbolTable()

for sym in ['<eps>', 'a', 'b']:
    input_sym.add_symbol(sym)

for sym in ['<eps>', 'd','c']:
    output_sym.add_symbol(sym)
```

## FST

```python
f = fst.Fst()
f.set_input_symbols(input_sym)
f.set_output_symbols(output_sym)

s0, s1, s2, s3 = f.add_state(), f.add_state(), f.add_state(), f.add_state()
```

```python
a = input_sym.find('a')
b = input_sym.find('b')
c = output_sym.find('c')
d = output_sym.find('d')
```
